```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("sales_train.csv")
```

```python
dataset.groupby('shop_id').sum()
```

```python
dataClean=dataset.groupby(['item_id','shop_id',"date_block_num"]).sum()
```

```python
dataClean
```

```python
dataClean.reset_index(inplace=True)
```

```python
sns.barplot(dataClean.shop_id, dataClean.item_cnt_day)
```

```python
dataClean
```

```python
dataClean.loc[dataClean["item_id"]==11373]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
X = dataClean[["item_id","shop_id"]]
y = dataClean.item_cnt_day
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
```

```python
linreg = LinearRegression()
linreg.fit(X_train, y_train)
```

```python
prediction = linreg.predict(X_test)
```

```python
np.sqrt(mean_squared_error(y_test, prediction))
```

```python
datatest = pd.read_csv("test.csv")
```

```python
predictfinal = linreg.predict(datatest[["item_id", "shop_id"]])
```

```python
dfFinal = pd.DataFrame({"ID" : datatest["ID"] , "item_cnt_month" : predictfinal})
```

```python
dfFinal.to_csv("FutureSalesPred.csv", index=False)
```
